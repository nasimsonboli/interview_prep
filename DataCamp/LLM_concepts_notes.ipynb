# The rise of LLMs in the AI Landscape
## AI and data-driven tasks
- AI tasks do well in sentiment analysis, fraud detection, etc. Still, they lacks the ability to understand context, respond to open-ended questions, and generate human-like responses in conversations.
- LLMs use deep learning techniques to perform a variety of NLP tasks such as text classification, summarization, generation, etc. They need large data and resources in order to work. They are powerful in processing and analyzing human language data. They can process unstructured data (multimodal data: text, audio, video, etc). They can do visual question answering (visual content, object identification, relationships, and scene description).
    -  they can do sentiment analysis, identify themes, translate text or speech, generate code, next word prediction, etc.
    -  **Challenges of LLMs**: Understanding the order of the words (word sequences), understanding the context in which the words are used, and connecting distant words (understanding long-range dependency) are challenging tasks for traditional LLMs.
    -  **Single-task Learning**: For example, question answering, text summarization, and language translation are trained for a specific task. These models are time and resource-intensive and can't use multi-modal data such as text, image, video, etc. They are less flexible and powerful compared to modern LLMs. 
    -  **Multi-task learning**: training a model on multiple tasks and data improves the flexibility of the model, but can decrease accuracy and efficiency. It also needs less data because the model learns from the data that is shared across tasks.
    - The novelty of LLMs is to detect language subtleties, like irony, humor, pun, sarcasm, intonation, intent and etc. They overcome the data's unstructured nature, and they outperform traditional NLP models. This is because of being trained on large training data and many parameters.

# LLM learning workflow

## 1. Text pre-processing 
Through the following steps, the text preprocessing turns unstructured data into a standardized format. These steps are independent and can be done separately depending on the task.
1. **Tokenization**: splits text into a **list** of words or tokens. Note that punctuation is also a token. e.g., ["The", "product", "is", "amazing"]
2. **Stop word removal**: removing frequent words that don't add meaning to the text.
3. **Lemmatization**: Reducing the words to their base form is called lemmatization. Then we group slightly different words with similar meaning.

The next step is to change the text into a form the computer understands. We do this through text representation, where preprocessed text turns into a numerical form.

## 2. Text representation
1. **Bag of words** approach involves converting the text into **a matrix of word counts**. 0 represents the absence of a word. This method has its own limitations:
   - including not being able to capture **the order of words or their context**, which can lead to incorrect interpretations of a text. As an example, the following two sentences mean differently despite sharing many words: "The cat chased the mouse swiftly", and "The mouse chased the cat."
    - BoW doesn't capture **the semantics between the words**. Therefore, it treats related words as independent. As an example, "cat" and "dog" are predator and prey.
3. **Word embeddings** address these limitations by capturing the semantic meanings of words. This is done by **allowing words to be represented as numbers, and therefore, similar words have similar representations**.
    - e.g., imagine (plant, carnivore, and furry) are features/embeddings learned by the model using the training data. Therefore, cat and mouse are each associated with a number for each animal. Higher numbers indicate stronger meanings. e.g., cat -> (-0.9, 0.9, 0.9) and mouse -> (-0.8, 0.7, -0.8). In reality, we don't see these labels for word embeddings, but instead we see only these lists of numbers as the embeddings. Since the embeddings are the same, the relationship between words is captured.

Note that 1 and 2 are data preparation worlflows.

## 3. Pre-training

- **Generative pre-training**: LLMs are typically trained using a technique called generative pre-training. It involves providing the model with input data of text tokens, and training it to predict the tokens within in. Two commonly used styles of generative pre-training are:

    - **Next word prediction** is a supervised learning technique that trains the model on the input data and its corresponding output (input-output pairs). A next word prediction model is used to train a language model to predict the next word in a sentence, given the context of the words before it. The model learns to generate coherent text by capturing the dependencies between words in the larger context. During training the model is presented with pairs of input and output examples. During training each generated output is added to the input for the next pair, allowing the model to predict the next output. e.g., 1) input: The quick brown, output: fox, 2) input: the quick brown fox, output: jumps. The model takes the second input to predict the "jumps" based on teh context of the previous input-output pair. The more examples it sees the better it predicts teh next word. Once trained, we can use the model to generate new sentences one word at a time.
    - **Masked language modeling** involves training a model to predict a masked word that is selectively hidden in a sentence. E.g., original text:"teh quick brown fox jumps over the lazy dog", Masked text:"The quick [MASK] fox jumps over the lazy dog". During training the model receives both the original and masked text as input. The model's objective is to correctly predict the missing word between "quick" and "fox". The model's objective is to correctly predict the missing word between "quick" and "fox". Even though the masked word could be any color, the model has learned that it is "brown" based on the training data.


- **Limitations of Pre-training**
    - "Largeness" of LLMs introduces some challenges, such as needing powerful computers and specialized infrastructure to train and deploy LLMs due to the massive amount of data and computational resources involved. Additionally, efficient model training methods and large amounts of high-quality training data are essential for optimal model performance. LLMs might need 100,000's CPUs and 10,000's GPUs compared to 4-8 CPU and 1-2 GPU for personal computers. 
    - Training and LLMs are time consuming (weeks, or months). Efficient model training allows models to train faster. On a single GPU training and LLM might take around 400 year.
    - They also need to be trained on high-quality data to learn the complexities and subtleties of language.

Fine-tuning addresses the limitations of pre-training by adapting a pre-trained model for specific tasks.

## 3.5. Addendum: Transformers

Transformers are part of pre-training. "Attention is all you need" revolutionized current LLMs. 

### Transformer architecture
- The transformer architecture emphasizes **long-range relationships between words** in a sentence to generate accurate and coherent text.
- 4 components:
    1) **text pre-processing** (tokenization, stop word removal, lemmatization) and **representation** (using word embeddings),
    2) **positional encoding** provides inforamtion about the position of the words in sequence, helping a transformer tie together distant words.
    3) **encoders** include the attention mechanism and neural network.
        - The **attention mechanism** directs attention to specific words and relationships. 
    5) **decoders** also use attention mechanism and NNs to process the encoded input and generate the final output.

The transofermer preprocesses input text, converts it into numbers  and incorporating position references . The encoder uses this information to encode the sentence, which the decoder then uses to predict subsequent words. the predicted word is added to the input, and the process continues until final output completes the input sentence.

### Transformers and long-range dependencies
Long-range dependencies require capturing relationships between distant words in a sentence which is challenging to model. The transformer's attention mechanism overcomes this limitation by focusing on different parts of the input. 
- E.g., "Jane who lives in NY and works as a software engineer, loves exploring new restaurants in the city." LLMs can attend to the relationship between the distant words "Jane" -- "loves exploring the city", leading to better contextual understanding.

### Process multiple parts simultaneously
When handling language, traditional LLMs are sequential, emaning that they process one words at a time. Transformers are an improvement in this area, because they focus on multiple parts of the input text simultaneously. This speeds up the process of understanding and generating text.

- E.g., "The cat sat on the mat" is processed by transformers to "cat", "sat", "on", "the", "mat" at the same time.

### Attention mechanism

- Attention mechanism help LLMs understand complex structures.
- They represent text more effectively by **focusing on important words and their relationships**.
- There are two types of attention mechanism:
    - **Self-attention** weighs the importance of each word **in a sentence** based on the context to capture long-range dependencies.
    - **Multi-headed attention** takes self-attention to the next level by splitting the input into multiple "heads" with each head focusing on different aspects of the relationship between words, allowing the model to learn a richer represenation of the text.
 
    - E.g., imagine a group conversation.
        - paying attention to the most relevant speaker, and filtering out the noise or less important comments, we can focus on the key points.
        - self-attention focuses on each person's words and evalute their relevance compared to eachother. So, the model weighs each speaker's input separately and combines them to form a more comprehensive understanding of the conversation.
        - multihead attention splits the attention into "multiple" channels that simultaneously focus on different aspects of the conversation. e.g. speaker's emotions, primary topic, related side-topics. Each aspect is processed independently and the results are merged to gain a more holistic perspective fo the conversation.
        - e.g. ""the boy went to buy some groceries, and he found discount on his favorite cereal"
            - attention: "boy", "store", "groceries", "discount"
            - self attention: "boy" and "he" are the same person.
            -  multihead: character("boy"), action("went to the store", "found a discount"), things involved ("groceries", "cereal")

## 4. Fine-tuning

So, pre-trained models from industry can be fine-tuned for specific tasks.
- They learn the **general language structure and flow** from large general-purpose datasets.
- Not optimized for specific tasks, but can be fine-tuned for a specific task.

### Fine-tuning vs pre-training
- Fine-tuning can be done with 1-2 CPU and a GPU compared to pretraining, which takes a thousand GPUs and CPUs.
- Fine-tuning takes hours to days for training time, compared to pretraining that takes weeks to months
- Fine-tuning takes around 1 gigabyte while pretraining takes hundreds of gigabytes.

### N-shot learning techniques

Training a pre-trained model for a specific task on a smaller dataset to improve performance.
N-shot learning is a kind of transfer learning which is learning from one task and transfers it to another related task.

- **Zero-shot learning**: no task-specific data. No explicit training. It uses its understanding of language and context to transfer knowledge to the new task. It can generalize without having seen prior examples.
- **Few-shot learning**: little task-specific data or a few examples. It uses the knowledge the model learned in the previous task to generalize to the new task with only a few examples.
- **Multi-shot learning**: relatively more training data than few-shot learning. It can save time in collecting and labeling  data while it has little to no compromise to accuracy.

## 5. Advanced Fine-tuning: Reinforcement Learning through Human Feedback

- LLM training: 1) pre-training, 2) fine-tuning, 3) Reinforcement Learning through Human Feedback (RLHF).
- why RLHF? The concern is that the large general-purpose training data may lack quality, meaning that it might contain noise,  errors, inconsistencies, reduced accuracy in specific tasks. E.g., imagine the model is trained on online forums and therefore, it contains unvalidated facts, and opinions. if the model treats this as training data, it can reduce model accuracy. RLHF introduces an external expert to validate the data and avoid these inaccuracies.
- pretrianing learns underlying language patterns and general language knowledge. However it doesn't capture context-specific complexities. During fine-tuning the model accuracy increases through using task-specific data and quality labeled data. This is where RLHF comes into play. RLHF techniques enhance fine-tuning through human feedback.

### Simplifying RLHF
- the model generates output, which is then reviewed by human who provides feedback on how well the model performed! the model is then updated based on the feedback to improve its performance over time.
    - Step 1: LLM receives a prompt and generates multiple responses to the given prompt. The responses are based on what it has learned rom reading lots of text.
    - Step 2: A human expert checkes the responses. The expert will rank the responses according to their quality, such as their accuracy, relevance, and coherence. This ranking process provides valuable informaiton to the model about whihc responses are better or worse.
    - Step3: the model learns from the expert's ranking to align its reponses in future with the expert's preferences.
